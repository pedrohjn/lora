# LoRa - Smart meter - setting from Georgiou and Raza "Low power wide..."

```python
import numpy as np
from scipy import special
from matplotlib import pyplot as plt
from matplotlib import rc, rcParams
#plt.rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
#plt.rc('text', usetex=True)
#Figures do not compile, because of dvipng. However the figures are saved right!
```

```python
#FIXED
# path loss exponent
alpha = 4
#Slotted Aloha prob.
p = 0.025
#SIR_th capture effect from nodes in the same SF 
beta1 = 4
#
#constant from PPP
k = np.pi * p * special.gamma(1 - 2/alpha) * special.gamma(1 + 2/alpha)
```

## Outage vs. Interference from same SF

```python
#Distance between node and BS
r = 1.5

#SIR_th from interferers with different SF
beta21 = 0.25
beta22 = 0.125
beta23 = 0.06
beta24 = 0.03
beta25 = 0.017
beta26 = 0.01
#
p1 = 1/6
p2 = 1/6
p3 = 1/6
p4 = 1/6
p5 = 1/6
p6 = 1/6
#
density_I = 0.05


plt.rc('text', usetex=True)
plt.rc('font', family='serif')
#Density of users in the same SF
density = np.linspace(0,4,300)

plt.plot(density, 1 - np.exp(- k * (r**2) * (beta1**(2/alpha) * p1 * density + beta21**(2/alpha) * density_I/p)),'k:',
            label = r'$\beta_\mathrm{SF7}=0.25$')
plt.plot(density, 1 - np.exp(- k * (r**2) * (beta1**(2/alpha) * p2 * density + beta22**(2/alpha) * density_I/p)),'g:',
             label = r'$\beta_\mathrm{SF8}=0.125$')
plt.plot(density, 1 - np.exp(- k * (r**2) * (beta1**(2/alpha) * p3 * density + beta23**(2/alpha) * density_I/p)),'b:',
             label = r'$\beta_\mathrm{SF9}=0.06$')
plt.plot(density,  1 - np.exp(- k * (r**2) * (beta1**(2/alpha) * p4 * density + beta24**(2/alpha) * density_I/p)),'k--',
             label = r'$\beta_\mathrm{SF10}=0.03$')
plt.plot(density,  1 - np.exp(- k * (r**2) * (beta1**(2/alpha) * p5 * density + beta25**(2/alpha) * density_I/p)),'g--',
             label = r'$\beta_\mathrm{SF11}=0.017$')
plt.plot(density,  1 - np.exp(- k * (r**2) * (beta1**(2/alpha) * p6 * density + beta26**(2/alpha) * density_I/p)),'b--',
             label = r'$\beta_\mathrm{SF12}=0.01$')
plt.plot(density,p1* (1 - np.exp(- k * (r**2) * (beta1**(2/alpha) * p1 * density + beta21**(2/alpha) * density_I/p))) +
                 p2* (1 - np.exp(- k * (r**2) * (beta1**(2/alpha) * p2 * density + beta22**(2/alpha) * density_I/p))) +
                 p3* (1 - np.exp(- k * (r**2) * (beta1**(2/alpha) * p3 * density + beta23**(2/alpha) * density_I/p))) +
                 p4* (1 - np.exp(- k * (r**2) * (beta1**(2/alpha) * p4 * density + beta24**(2/alpha) * density_I/p))) +
                 p5* (1 - np.exp(- k * (r**2) * (beta1**(2/alpha) * p5 * density + beta25**(2/alpha) * density_I/p))) +
                 p6* (1 - np.exp(- k * (r**2) * (beta1**(2/alpha) * p6 * density + beta26**(2/alpha) * density_I/p))) 
        ,'r-',label = 'Average')


plt.axis([0,4,0,1])
plt.xlabel('$\lambda_\mathrm{SF}$', fontsize=12)
plt.ylabel(r'Outage probability', fontsize=12)
#plt.legend(bbox_to_anchor=(1.,1),loc=2,fontsize=12)
plt.legend(loc=2,fontsize=10)
plt.grid(color='k', alpha=0.5, linestyle='dashed', linewidth=0.5)


plt.savefig('out-lsf.pdf', bbox_inches='tight')
#plt.show()
```
